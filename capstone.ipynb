```python
import pandas as pd
```

### Gathering Data

```python
#Menampilkan tabel store

data = pd.read_csv('store.csv')
data.head()
```

### Assessing Data

```python
data.info()
```

```python
#Mengecek missing value

data.isnull().sum()
```

```python
#Mengecek data duplicate

data.duplicated().sum()
```

```python
#Mengecek data yang tidak sesuai

data.describe()
```

### Cleaning Data

```python
#Membersihkan data duplikat

data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.info()
```

### Exploratory Data

```python
data.describe(include='all')
```

```python
#Mengecek kota yang memiliki penjualan terbanyak

data.groupby(by="City").Customer_ID.nunique().sort_values(ascending=False).head(10)
```

```python
#Mengecek negara yang memiliki penjualan terbanyak

data.groupby(by="State").Customer_ID.nunique().sort_values(ascending=False).head(10)
```

```python
#Mengecek pendapatan penjualan

data["Revenue"] = data.Sales*data.Quantity

data.groupby(by="Product_Name").agg({
    "Revenue": "sum"
}).sort_values(by="Revenue", ascending=False).head(10)
```

```python
#Mengecek profit penjualan

data.groupby(by="Ship_Date").agg({
    "Profit": "sum"
}).sort_values(by="Profit", ascending=False).head(10)
```

```python
#Produk yang paling diminati dan kurang diminati

data.groupby(by="Product_Name").agg({
    "Category": "min",
    "Quantity": "sum"
}).sort_values(by="Quantity", ascending=False).head(5)
```

```python
#Produk kurang diminati

data.groupby(by="Product_Name").agg({
    "Category": "min",
    "Quantity": "sum"
}).sort_values(by="Quantity", ascending=True).head(5)
```

```python
data.to_csv("fix_store.csv", index=False)
```
